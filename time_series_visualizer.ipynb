```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('fcc-forum-pageviews.csv', index_col = ['date'])
```

```python
mask_value = (df['value'] < df['value'].quantile(0.975)) & (df['value'] > df['value'].quantile(0.025))
df = df[mask_value]
```

```python
df.plot(figsize=(20, 6),title = 'Daily freeCodeCamp Forum Page Views 5/2016-12/2019', xlabel='Date', ylabel='Page Views', color='red')
```
